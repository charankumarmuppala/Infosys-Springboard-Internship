# Netflix Dataset - Data Cleaning

Handling null values, duplicates, inconsistent columns, and date formatting.

```
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
df.shape
```

### Check Null Values & Data Types

```
df.isnull().sum()
```

```
df.dtypes
```

### 1. Fix Misplaced Values (Duration stuck in Rating column)

```
# 3 rows have duration values like '74 min' in the rating column
misplaced = df['rating'].str.contains('min', na=False)
df.loc[misplaced, 'duration'] = df.loc[misplaced, 'rating']
df.loc[misplaced, 'rating'] = pd.NA
```

### 2. Fill Null Values with 'Not Available'

```
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna('Not Available')
```

### 3. Drop Rows where Rating, Duration, Date_added are Null

```
df.dropna(subset=['rating', 'duration', 'date_added'], inplace=True)
print(f"Shape after dropping nulls: {df.shape}")
```

### 4. Remove Duplicates

```
print(f"Duplicates found: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
```

### 5. Convert date_added to dd-mm-yyyy Format

```
df['date_added'] = pd.to_datetime(df['date_added'].str.strip()).dt.strftime('%d-%m-%Y')
df['date_added'].head()
```

### 6. Verify Cleaning

```
print(f"Final Shape: {df.shape}")
print(f"\nNull Values:\n{df.isnull().sum()}")
df.head()
```

### 7. Save Cleaned Dataset

```
df.to_csv('netflix_cleaned.csv', index=False)
print("✅ Saved successfully!")
```
